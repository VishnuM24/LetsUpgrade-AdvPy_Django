Convert 15-08-2020 15:20:20 into **epoch time**

##***SOLUTION***
---


```
import time, calendar  # calendar module for GMT time

def HR_EP( timeString, gmt= True ):
  t9 = time.strptime( timeString, "%d-%m-%Y %H:%M:%S" )
  if gmt :
    ep_time = calendar.timegm( t9 ) # GMT TIMEZONE
  else:
    ep_time = time.mktime( t9 )  # LOCAL TIMEZONE
  return ep_time

if __name__ == '__main__':
  time1 = input( 'Enter date and time [ DD-MM-YYYY HH:MM:SS ] :' )

  print( 'Local Epoch Time :' , HR_EP( time1, gmt= False ) )
  print( 'GMT Epoch Time :' , HR_EP( time1, gmt= True ) )
```
